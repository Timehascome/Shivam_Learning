```python
# 1) Cinema ticket price
age = int(input("Enter your age: "))
if age >= 18:
    print("Ticket Price: ₹200")
else:
    print("Ticket Price: ₹150")
```

```python
# 2) Pass/Fail
marks = float(input("Enter your marks: "))
if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
# 3) Discount
amount = float(input("Enter purchase amount: "))
if amount >= 5000:
    print("20% Discount")
else:
    print("No Discount")
```

```python
# 4) Voting eligibility
age = int(input("Enter your age: "))
if age >= 18:
    print("Eligible to Vote")
else:
    print("Not Eligible to Vote")
```

```python
# 5) Loan eligibility
age = int(input("Enter your age: "))
salary = float(input("Enter your salary: "))
if age >= 21 and salary >= 25000:
    print("Eligible for Loan")
else:
    print("Not Eligible for Loan")
```

```python
# 6) Password strength check
pwd = input("Enter password: ")
specials = "!@#$%^&*"
has_digit = any(ch.isdigit() for ch in pwd)
has_upper = any(ch.isupper() for ch in pwd)
has_lower = any(ch.islower() for ch in pwd)
has_special = any(ch in specials for ch in pwd)
if len(pwd) > 8 and has_digit and has_upper and has_lower and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
# 7) Free delivery
order_amount = float(input("Enter order amount: "))
premium = input("Are you a premium member? (yes/no): ").strip().lower()
is_premium = premium in ("yes", "y")
if order_amount >= 1000 or is_premium:
    print("Free Delivery")
else:
    print("Delivery Charges Apply")
```
