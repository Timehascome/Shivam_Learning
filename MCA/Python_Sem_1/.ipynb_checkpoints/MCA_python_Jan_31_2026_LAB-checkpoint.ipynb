```python
# 1) Shopping Cart Calculations
prices = [25.99, 15.50, 8.75]
tax_rate = 0.08

subtotal = sum(prices)
tax = subtotal * tax_rate
total_cost = subtotal + tax

print("1) Subtotal:", round(subtotal, 2))
print("   Tax:", round(tax, 2))
print("   Total:", round(total_cost, 2))


# 2) Temperature Conversion
c = 25
f_from_c = (c * 9/5) + 32

f = 77
c_from_f = (f - 32) * 5/9

print("\n2) 25°C ->", round(f_from_c, 1), "°F")
print("   77°F ->", round(c_from_f, 1), "°C")


# 3) Recipe Scaling (multiply each ingredient by 4)
flour_cups = 2
sugar_cups = 1
eggs = 2
milk_cups = 0.5

scale = 4
print("\n3) Scaled Recipe (x4):")
print("   Flour (cups):", flour_cups * scale)
print("   Sugar (cups):", sugar_cups * scale)
print("   Eggs:", eggs * scale)
print("   Milk (cups):", milk_cups * scale)


# 4) Student Grade Calculation
quizzes = [85, 92, 78]
midterm = 88
final_exam = 95

quiz_avg = sum(quizzes) / len(quizzes)
overall = (0.30 * quiz_avg) + (0.30 * midterm) + (0.40 * final_exam)

print("\n4) Quiz Average:", round(quiz_avg, 2))
print("   Overall Grade:", round(overall, 2))


# 5) Travel Distance and Time
distance_km = 350
speed_kmh = 80

hours_decimal = distance_km / speed_kmh
hours_int = int(hours_decimal)
minutes = (hours_decimal - hours_int) * 60

print("\n5) Travel time (hours decimal):", round(hours_decimal, 3))
print("   Travel time:", hours_int, "hours and", round(minutes, 1), "minutes")


# 6) Inventory Management
starting_inventory = 150
sold = 87
returned = 5
new_shipment = 50

current_stock = starting_inventory - sold + returned + new_shipment

print("\n6) Current stock level:", current_stock)


# 7) Cooking Measurements Conversion
tablespoons = 3
teaspoons = tablespoons * 3  # 1 tbsp = 3 tsp

cups = 2
fluid_ounces = cups * 8      # 1 cup = 8 fl oz

print("\n7) 3 tbsp ->", teaspoons, "tsp")
print("   2 cups ->", fluid_ounces, "fl oz")


# 8) Sports Statistics (per game averages)
games = 82
total_points = 2015
total_rebounds = 845
total_assists = 512

ppg = total_points / games
rpg = total_rebounds / games
apg = total_assists / games

print("\n8) Points per game:", round(ppg, 2))
print("   Rebounds per game:", round(rpg, 2))
print("   Assists per game:", round(apg, 2))


# 9) Weight Conversion (round to 1 decimal)
lb_to_kg_factor = 0.453592

lbs = 180
kg = lbs * lb_to_kg_factor

kg2 = 75
lbs2 = kg2 / lb_to_kg_factor

print("\n9) 180 lb ->", round(kg, 1), "kg")
print("   75 kg ->", round(lbs2, 1), "lb")


# 10) Photo Resize + Aspect Ratio
orig_w, orig_h = 4000, 3000
scale = 0.25

new_w = int(orig_w * scale)
new_h = int(orig_h * scale)

aspect_ratio = orig_w / orig_h

print("\n10) Resized dimensions:", new_w, "x", new_h)
print("    Aspect ratio (w/h):", round(aspect_ratio, 4))
```
