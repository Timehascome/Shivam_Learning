# **Tuple**

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

````python
# Creating a Tuple

# Empty tuple
t1 = ()
print(t1)

# Tuple with elements
t2 = (1, 2, 3, 4)
print(t2)

# Mixed data types
t3 = (10, "Python", 3.14, True)
print(t3)

# Single element tuple (important comma)
t4 = (5,)
print(t4)
````

````python
t = (1, 2, 3, "Hello", [4, 8, 16],(1,))
print(type(t))
````

````python
# Accessing Items

t = (10, 20, 30, 40)

# Indexing
print(t[0])      # 10
print(t[-1])     # 40

# Slicing
print(t[1:3])    # (20, 30)
````

````python
# Editing Items

t = (1, 2, 3)
t[0] = 10   # TypeError
````

```
But we can convert it into a list, modify it, and convert back. similarly for Adding Items, Deleting Items
```

````python
t = (1, 2, 3)
temp = list(t)
temp[0] = 10
t = tuple(temp)
print(t)   # (10, 2, 3)
````

````python
# But we can create a new tuple by concatenation:

t = (1, 2, 3)
t = t + (4,)
print(t)   # (1, 2, 3, 4)
````

````python
# But we can delete the entire tuple:

t = (1, 2, 3)
del t
````

````python
# Operations on Tuples

t1 = (1, 2, 3)
t2 = (4, 5)

# Concatenation
print(t1 + t2)      # (1, 2, 3, 4, 5)

# Repetition
print(t1 * 2)       # (1, 2, 3, 1, 2, 3)

# Membership
print(2 in t1)      # True

# Length
print(len(t1))      # 3
````

````python
# Tuple Functions Example

t = (10, 20, 30, 20)

# 1. Count occurrences of an element
print("Count of 20:", t.count(20))   # 2

# 2. Find index of first occurrence
print("Index of 30:", t.index(30))   # 2

# 3. Length of tuple
print("Length:", len(t))             # 4

# 4. Minimum value
print("Minimum:", min(t))            # 10

# 5. Maximum value
print("Maximum:", max(t))            # 30

# 6. Sum of elements
print("Sum:", sum(t))                # 80

# 7. Sorted tuple (returns list)
print("Sorted:", sorted(t))          # [10, 20, 20, 30]

# 8. Check membership
print("Is 10 in tuple?", 10 in t)    # True

# 9. Reverse tuple
print("Reversed:", tuple(reversed(t)))  # (20, 30, 20, 10)
````

**Problem 1: GPS Location System**

**Scenario:**
A travel mapping app uses tuples to store GPS coordinates of different landmarks to ensure data integrity — once stored, the coordinates cannot be modified directly (immutability keeps records authentic).



````python
# Locations: (latitude, longitude, landmark_name, city)
locations = [
    (28.6139, 77.2090, "India Gate", "New Delhi"),
    (19.0760, 72.8777, "Gateway of India", "Mumbai"),
    (13.0827, 80.2707, "Marina Beach", "Chennai"),
    (34.0522, -118.2437, "Hollywood Sign", "Los Angeles"),
    (40.7128, -74.0060, "Statue of Liberty", "New York")
]
````

**Task A:** Location Dictionary

Create a dictionary where tuple coordinates are keys and landmark names are values.


**Task B:** Update Coordinates

It’s discovered that the stored coordinates for Marina Beach are slightly off. The correct coordinates are (13.0500, 80.2824).
Create a new locations list with this correction — don’t modify the existing tuples directly, since tuples are immutable.


**Task C:** Find Nearby Locations

FindNearby location based on Latitude.
1. Southern India: 12.0°N to 20.0°N
2. Northern India: 25.0°N to 35.0°N
3. USA: 30.0°N to 45.0°N'

Example:
find_locations_in_latitude_range(locations, 12.0, 20.0)
→ should return all landmarks located in southern India.

**Task D:** Add New Location

Add “Taj Mahal” in Agra with coordinates (27.1751, 78.0421) to the locations list — again, do this without modifying existing tuples, maintaining immutability.

````python
dict_map={}
for lat, long, landmark, city in locations:
    dict_map[(lat, long)] = landmark
print(dict_map)
````

````python
new_locations = [
    (13.0500, 80.2824, name, city) if name == "Marina Beach" else (lat, lon, name, city)
    for lat, lon, name, city in locations
]
print(new_locations)

new_locations.append((27.1751, 78.0421, "Taj Mahal", "Agra"))

print(new_locations)
````

````python
def find_locations_in_latitude_range(locations, min_lat, max_lat):
    return [
        (lat, lon, name, city)
        for lat, lon, name, city in locations
        if min_lat <= lat <= max_lat
    ]

ans =find_locations_in_latitude_range(locations, 12.0, 20.0)
print(ans)
````
