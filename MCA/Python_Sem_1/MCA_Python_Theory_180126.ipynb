```python
a = 5
b = 7

c = a   # c = 5
a = b   # a = 7
b = c   # b = 5

print(a, b)

# outcome:
# a = 7
# b = 5
# swapping using 3rd or temporary variable
```

```python
a = 5
b = 7
print(a, b)


a = a+b  # a = 5+7 = 12
b = a-b  # b = 12-7 = 5
a = a-b  # a = 12-5 = 7

print(a, b)
#swapping 2 no.s without using 3rd variable
```

```python
a = 5
b = 7

print(a,b)

(a, b) = (b, a)

print(a,b)
```

```python
name = 'AI'
s = f"{name:*^10}"
print(s)

# outcome: 10 character length
# fill character: *
# ^ Center alignment
```

```python
name = 'AI'
s = f"{name:+^10}"
print(s)
```

```python
name = 'AI'
s = f"{name:*>10}"
print(s)
# > right alignment
```

```python
name = 'AI'
s = f"{name:*<10}"
print(s)
# < left alignment
```
