```python
# dim_movie:
# movie_id (int)
# title (string)
# release_year (int, parsed from title like “(1995)”)
# imdb_id, tmdb_id
# genres as:
# either genre_list (python list as string)
# AND/OR a bridge table (below)

dim_movie = df_movies.merge(df_links, on ="movieId", how='inner')
dim_movie['release_year']=dim_movie['title'].str.split().str[-1].replace('[()]','', regex=True)
dim_movie['title'] = dim_movie['title'].str.split().str[:-1].str.join(' ')
dim_movie['genres']=dim_movie['genres'].str.split("|")
print(dim_movie.head())
bridge_movie_genre = dim_movie[['movieId', 'genres']]
bridge_movie_genre = (dim_movie[['movieId', 'genres']].explode('genres').rename(columns={'genres': 'genre'}))
bridge_movie_genre.head()


```
