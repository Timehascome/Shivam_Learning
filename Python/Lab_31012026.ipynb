```python
# Complex Number Division

c1 = 6 + 8j
c2 = 2 + 2j

# Formula: (a + bj) / (c + dj) = [(ac + bd) / (c² + d²)] + [(bc - ad) / (c² + d²)] j
# Here: (6+8j) / (2+2j) = [(6*2 + 8*2) / (2²+2²)] + [(8*2 - 6*2) / (2²+2²)] j
#      = (12+16)/8 + (16-12)/8 j
#      = 28/8 + 4/8 j = 3.5 + 0.5j
result = c1 / c2
print(f"Complex division: {c1} / {c2} = {result}")

# Complex divided by Real
# Formula: (a + bj) / k = (a/k) + (b/k) j
# Here: (6+8j)/2 = 6/2 + 8/2 j = 3 + 4j
result2 = c1 / 2
print(f"Complex by real: {c1} / 2 = {result2}")
```

```python
# Equal to
print(5 == 5)
print(5 == 3)

# Not equal to
print(5 != 3)
print(5 != 5)

# Greater than
print(5 > 3)
print(3 > 5)

# Less than
print(3 < 5)
print(5 < 3)

# Greater than or equal to
print(5 >= 5)
print(5 >= 3)
print(3 >= 5)

# Less than or equal to
print(3 <= 3)
print(3 <= 5)
print(5 <= 3)
```

# Logical Operators

| Operator | Name          | Description / Works With            
|----------|---------------|-----------------------------------
| `and`    | Logical AND    | Returns True if **both operands** are True; works with bool expressions
| `or`     | Logical OR     | Returns True if **at least one operand** is True; works with bool expressions
| `not`    | Logical NOT    | Reverses the Boolean value; works with a single bool expression


```python
# Basic AND operations
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
# Short-circuit behavior of "and":
# In Python, the "and" operator evaluates expressions left to right.
# - It stops evaluating as soon as it finds a False (falsy) value (short-circuit).
# - If all values are True (truthy), it returns the last evaluated value.

print(0 and 5)
# 0 → The first value (0) is false, so Python stops here and returns 0.

print(3 and 0)
# 0 → The first value (3) is true, so Python evaluates the second value (0).

print(2 and 3)
# 3 → Both values are true, so Python returns the **last value** (3).

print("" and "hello")
# "" → The first value is an empty string, which is falsy.
# Python stops immediately and returns "", without evaluating "hello".

print("Hi" and "hello")
```

```python
name = "Alice"
age = 25

result = name and age >= 18 # >= has higher precedence in python
print(result)
```

```python
# Basic OR operations
print(True or True)     # True
print(True or False)    # True
print(False or True)    # True
print(False or False)   # False
```

```python
# In Python, the "or" operator evaluates expressions left to right.
# - It stops evaluating as soon as it finds a True (truthy) value.
# - If all values are False (falsy), it returns the last evaluated value.

print(0 or 5)
# 5 → The first value (0) is falsy, so Python evaluates the second value (5).
# 5 is true, so it returns 5 immediately.

print(3 or 0)
# returns 3 immediately without checking 0.

print(0 or "" or None or 7)
print(0 or "" or None)
```

```python
username = "" or "guest"
print(username)

config = {}
port = config.get('port') or 8080
print(port)

def get_data():
    return None

data = get_data() or [] # Both are falsy, but last operand
print(data)
```

```python
# Basic NOT operations
print(not True)         # False
print(not False)        # True

# With various data types
print(not 0)            # True
print(not 1)            # False
print(not "")           # True
print(not "hello")      # False
print(not [])           # True
print(not [1, 2])       # False

# Practical examples
# Input validation
user_input = ""
if not user_input:
    print("Input cannot be empty")

# Check if list is empty
items = []
if not items:
    print("No items found")

# Toggle boolean
is_active = True
is_active = not is_active
print(is_active)  # False
```

```python
# Customer purchases
item1_price = 25.99
item2_price = 15.50
item3_price = 8.75
tax_rate = 0.08  # 8%

# Calculate subtotal
subtotal = item1_price + item2_price + item3_price
# Calculate tax amount
tax_amount = subtotal * tax_rate
# Calculate total
total = subtotal + tax_amount

print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax: ${tax_amount:.2f}")
print(f"Total: ${total:.2f}")
```

```python
# Convert Celsius to Fahrenheit
celsius = 25
fahrenheit = (celsius * 9/5) + 32

print(f"{celsius}°C = {fahrenheit}°F")

# Convert Fahrenheit to Celsius
fahrenheit = 77
celsius = (fahrenheit - 32) * 5/9

print(f"{fahrenheit}°F = {celsius:.1f}°C")
```

```python
# Trip planning
distance_km = 350  # kilometers
speed_kmh = 80     # kilometers per hour

# Calculate travel time
travel_time_hours = distance_km / speed_kmh
# Convert to hours and minutes
hours = int(travel_time_hours)
minutes = (travel_time_hours - hours) * 60

print(f"Distance: {distance_km} km")
print(f"Speed: {speed_kmh} km/h")
print(f"Estimated travel time: {hours} hours {minutes:.0f} minutes")
```

```python
# Basketball player stats
games_played = 82
points_total = 2015
rebounds_total = 845
assists_total = 512

# Calculate averages per game
points_per_game = points_total / games_played
rebounds_per_game = rebounds_total / games_played
assists_per_game = assists_total / games_played

print(f"Games played: {games_played}")
print(f"Points per game: {points_per_game:.1f}")
print(f"Rebounds per game: {rebounds_per_game:.1f}")
print(f"Assists per game: {assists_per_game:.1f}")
```

```python
# Room dimensions for flooring
room_length = 4.5  # meters
room_width = 3.2   # meters
tile_size = 0.6    # square meters per tile

# Calculate area and number of tiles needed
room_area = room_length * room_width
tiles_needed = room_area / tile_size
# Add 10% extra for waste
extra_tiles = tiles_needed * 0.10
total_tiles = tiles_needed + extra_tiles

print(f"Room area: {room_area:.2f} square meters")
print(f"Tiles needed: {tiles_needed:.1f}")
print(f"Extra (10%): {extra_tiles:.1f}")
print(f"Total tiles to buy: {total_tiles:.0f}")
```

```python
# Household electricity usage
kwh_rate = 0.12  # $ per kWh
daily_usage = 15  # kWh per day
days_in_month = 30

# Calculate monthly cost
monthly_usage = daily_usage * days_in_month
monthly_cost = monthly_usage * kwh_rate

print(f"Daily usage: {daily_usage} kWh")
print(f"Monthly usage: {monthly_usage} kWh")
print(f"Monthly cost: ${monthly_cost:.2f}")
```
