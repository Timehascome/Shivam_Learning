```python
# we try the functools module
import functools    
@functools.lru_cache(maxsize=4)
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)    
print(fib(10))

#we try other features of functools
def decorator_func(original_func):
    @functools.wraps(original_func)
    def wrapper_func(*args,**kwargs):
        print(f"Wrapper executed before {original_func.__name__}")
        return original_func(*args,**kwargs)
    return wrapper_func
@decorator_func
def display_info(name,age):
    print(f"display_info ran with arguments ({name},{age})")
display_info("Alice",30)  

```

```python
#we now try the dundermethods
class Sample:
    def __init__(self,value):
        self.value=value
    def __add__(self,other):
        return Sample(self.value+other.value)
    def __str__(self):
        return f"Sample with value {self.value}"    
s1=Sample(10)
s2=Sample(20)   
s3=s1+s2
print(s3)   

#what is the above code doing?
#1. When we do s1+s2, Python internally calls the __add__ method of the Sample class.
#2. The __add__ method takes another Sample object as an argument (other) and   returns a new Sample object with the sum of the values of the two objects.
#3. Finally, when we print s3, Python calls the __str__ method to get a string representation of the object, which is then printed to the console.
```
