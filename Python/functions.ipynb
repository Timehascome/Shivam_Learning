```python
#functions using map
list_a = [1,2,3,6]

list_b = [x**2 for x in list_a]
print(list_b)

def func(x):return x**2  
list_c = list(map(func, list_a))
print(list_c)

list_d = list(map(lambda x : x**2, list_a))
print(list_d)
```

```python
#functions using filter
list_a = [1,2,3,6]
def func(x):return x**2
list_e = list(filter(lambda x : x >5, map(func, list_a)))
print(list_e)
```

```python
#functions using lambda and list comprehension
people = [
    {'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 35},
    {'name': 'Charlie', 'age': 40},
    {'name': 'Dave', 'age': 25},
]
list_name = list([x['name'] for x in people if x['age']>30])
print(list_name)

list_name_a = list(map(lambda x : x['name'],filter(lambda x: x['age']>30, people)))
print(list_name_a)
```

```python
#understanding *args, **kwargs
def list_agg(*args):                               # example of args
    for i in args:
        print(i)
    return sum(i)/len(i)    

list_a = [1,2,3,4]
print(list_agg(list_a))

def dict_agg(**kwargs):                              #example of kwargs
    for i,j in kwargs.items():
        print(f"{i} : {j}")
        
dict_agg(name= 'Alice',age= 30)

def fun(*args,**kwargs):                             #using both
    print(args, kwargs) 
fun(3,4)
```

```python
#decorator # generator
import functools
def memory(func):
    def inner(*args,**kwargs):        
        data = func(*args)
        return data*2
    return inner

@memory                                             # example of a decorator
def add(x,y):
    return x+y

print(f"Example of a decorator {add(3,4)}")


def gen(x):
    for i in range(0,x):
        yield i                                      #generator
list_gen = gen(5)
for i in list_gen:
    print(f"This is from generator yield")
    print(i)
```

```python
#logging
import logging
logging.basicConfig(level=logging.ERROR)
def calculate_statistics(*args, **kwargs):
    try:
        numbers=[]
        numbers.extend(args)
        numbers.extend(kwargs.values())
        numbers=[float(x) for x in numbers]
        print(numbers)
    except ValueError as e:
        logging .error(f"ValuerroR:{e}")
        return {'key': 'NoValues'}
    except exception as e:
        logging.error(e)
        return {'key': 'NoValues'}
    return {
        'sum' : sum(numbers),
        'min' : min(numbers),
        'max' : max(numbers),
        'avg' : sum(numbers)/len(numbers)
    }    
result_a=calculate_statistics(10,20,30,40)
print(result_a)
result_b=calculate_statistics(a=1,b=2,c=3)
print(result_b)
result_c=calculate_statistics("abc",1,2)   
print(result_c)
result_d = calculate_statistics(10,20,a=30,b=400)
print(result_d)
```

```python
#context manager
import os
with open('test.txt','w') as f:
    f.write("This is a test file to understand contextmanager in python.")
with open('test.txt','r') as f:
    content = f.read()
    print(content)  

print(f.closed)
print(os.getcwd())    #current working directory

#file is closed outside the with block

from contextlib import contextmanager
@contextmanager
def change_dir(destination):
    try:
        current_dir = os.getcwd()
        os.chdir(destination)
        yield
    finally:
        os.chdir(current_dir)     

with change_dir('c:\\Users\\399sh\\Desktop\\git\\Shivam_learning\\'):
    print(f"Changed directory to {os.getcwd()}")
print(f"Back to the original directory {os.getcwd()}")


```
