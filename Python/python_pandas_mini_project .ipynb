```python
print(df['gender'].value_counts())
print(df.groupby(df['gender']).count())
print(df.groupby('gender').size())
print(df['gender'].value_counts()['M'] / len(df) * 100)
print((df['gender'].value_counts(normalize=True) * 100).round(2))
```

```python
import pandas as pd

list_1 = [1,2,3,4,5]
list_2 = [1,6,7,8,9,10]

df_1 = pd.DataFrame(list_1, columns=['col_A'])
df_2 = pd.DataFrame(list_2, columns=['col_B'])
```

```python
pd.merge(df_1, df_2, left_on='col_A', right_on='col_B', how='inner')
```

```python
pd.merge(df_1, df_2, left_on='col_A', right_on='col_B', how='left')
```

```python
pd.merge(df_1, df_2, left_on='col_A', right_on='col_B', how='right')
```

```python
pd.merge(df_1, df_2, left_on='col_A', right_on='col_B', how='outer')
```

```python
df_1.join(df_2, how='right')
```
